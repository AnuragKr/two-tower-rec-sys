# Content Data Cleaning

```python
#loading user csv data

# Read CSV using the python engine with proper quoting
df_content = pd.read_csv(
    "/content/drive/MyDrive/Companies Interview Projects/InShorts/data/training_content/part-00000-a34a1545-5cf1-47b9-93c2-29c1d3f0bfb7-c000.csv",
    engine='python',        # python engine handles quotes and newlines better
    quoting=csv.QUOTE_ALL,  # treat text in quotes as one field
    quotechar='"',          # standard quote char
    escapechar='\\',        # handle escaped quotes
    on_bad_lines='warn'     # just warn for truly broken lines
)

#preview the head data
df_content.head(5)
```

```python
df_content.columns
```

```python
df_content.dtypes
```

```python
df_content.info()
```

```python
df_content.shape
```

```python
df_content.describe()
```

```python
df_content.isnull().sum()
```

```python
#removing the duplicates hashid
df_content.drop_duplicates(subset='hashid', inplace=True)
```

```python
df_content['newsType'].value_counts()
```

```python
df_content['categories'].value_counts()
```

```python
df_content['hashtags'].value_counts()
```

```python
df_content['newsDistrict'].value_counts()
```

```python
#analyzing content length for text news and video news
df_content['content_length'] = df_content['content'].apply(lambda x: len(str(x)))
text_news = df_content[df_content['newsType'] == 'NEWS']
video_news = df_content[df_content['newsType'] == 'VIDEO_NEWS']
print(text_news['content_length'].mean(), video_news['content_length'].mean())
```

```python
#Filling News Type
def infer_news_type(row):
    text = str(row['content']).lower()
    if len(text.split()) < 70 or any(word in text for word in ['watch', 'video', 'clip', 'see']):
        return 'VIDEO_NEWS'
    elif str(row['categories']).lower() in ['entertainment', 'travel', 'hatke']:
        return 'VIDEO_NEWS'
    else:
        return 'NEWS'

df_content['newsType'] = df_content['newsType'].fillna('Unknown_Type')
mask_unknown = df_content['newsType'].isin(['Unknown_Type'])
df_content.loc[mask_unknown, 'newsType'] = df_content[mask_unknown].apply(infer_news_type, axis=1)
```

```python
#setting null values to Unknown_Author
df_content['author'] = df_content['author'].fillna('Unknown_Author')
```

```python
#setting source name to Unknown_Source
df_content['sourceName'] = df_content['sourceName'].fillna('Unknown_Source')
```

```python
df_content.isnull().sum()
```

```python
df_content = df_content.drop(columns=['newsDistrict'])
```

```python
df_content["newsLanguage"].value_counts()
```

```python
from langdetect import detect, DetectorFactory
DetectorFactory.seed = 0  # ensures consistent results

def detect_language_safe(text):
    """Detect language safely; return 'unknown' if detection fails."""
    try:
        return detect(str(text))
    except:
        return 'unknown'
```

```python
mask_missing_lang = df_content['newsLanguage'].isna() | (df_content['newsLanguage'] == '')

# Detect language for missing ones
df_content.loc[mask_missing_lang, 'detected_lang'] = df_content.loc[mask_missing_lang, 'content'].apply(detect_language_safe)
```

```python
df_content['detected_lang'].value_counts()
```

```python
df_content.loc[
    (mask_missing_lang) & (df_content['detected_lang'] == 'en'),
    'newsLanguage'
] = 'english'

df_content.loc[
    (mask_missing_lang) & (df_content['detected_lang'] != 'en'),
    'newsLanguage'
] = 'Unknown_Language'
```

```python
df_content['newsLanguage'].value_counts()
```

```python
df_content.drop(columns=['detected_lang'], inplace=True)
```

```python
df_content.isnull().sum()
```

```python
#Handling of hastag values
df_content['hashtags'] = df_content['hashtags'].fillna('')

df_content['hashtags'] = df_content['hashtags'].astype(str).str.replace('#', '', regex=False)
df_content['hashtags'] = df_content['hashtags'].str.replace(',', ' ', regex=False)
df_content['hashtags'] = df_content['hashtags'].str.replace('_', ' ', regex=False)
```

```python
df_content['title'] = ( df_content['title'].fillna('') + ' ' + df_content['hashtags'].fillna(''))
df_content.drop(columns=['hashtags'], inplace=True)
```

```python
df_content.isnull().sum()
```

```python
df_content.to_csv('/content/drive/MyDrive/Companies Interview Projects/InShorts/cleaned_data/content_cleaned_data.csv', index=False)
```

## Interaction Data Loading

```python
#loading user csv data

# Read CSV using the python engine with proper quoting
df_interaction_1 = pd.read_csv("/content/drive/MyDrive/Companies Interview Projects/InShorts/data/event/part-00000-7e210b01-29d8-430f-988f-d3e3b34da614-c000.csv")
df_interaction_2 = pd.read_csv("/content/drive/MyDrive/Companies Interview Projects/InShorts/data/event/part-00001-7e210b01-29d8-430f-988f-d3e3b34da614-c000.csv")
df_interaction_3 = pd.read_csv("/content/drive/MyDrive/Companies Interview Projects/InShorts/data/event/part-00002-7e210b01-29d8-430f-988f-d3e3b34da614-c000.csv")
df_interaction_4 = pd.read_csv("/content/drive/MyDrive/Companies Interview Projects/InShorts/data/event/part-00003-7e210b01-29d8-430f-988f-d3e3b34da614-c000.csv")

df_interaction = pd.concat([df_interaction_1,df_interaction_2,df_interaction_3,df_interaction_4], ignore_index=True)

#preview the head data
df_interaction.head(5)
```

```python
df_interaction.shape
```

```python
df_interaction.isnull().sum()
```

```python
#dropping hashid column
df_interaction = df_interaction.dropna(subset=['hashId'])
```

```python
df_interaction.isnull().sum()
```

```python
df_interaction["categoryWhenEventHappened"].value_counts()
```

```python
#setting null value to Unknown for categoryWhenEventHappened
df_interaction['categoryWhenEventHappened'] = df_interaction['categoryWhenEventHappened'].fillna('Unknown')
```

```python
#setting cardview position to median
df_interaction['cardViewPosition'] = pd.to_numeric(df_interaction['cardViewPosition'], errors='coerce').fillna(df_interaction['cardViewPosition'].median())
```

```python
#setting overallTimeSpent to 0
df_interaction['overallTimeSpent'] = pd.to_numeric(df_interaction['overallTimeSpent'], errors='coerce').fillna(0)
```

```python
df_interaction["searchTerm"].value_counts()
```

```python
df_interaction["relevancy_color"].value_counts()
```

```python
df_interaction["relevancy_topic"].value_counts()
```

```python
#handling search term
df_interaction['searchTerm'] = (df_interaction['searchTerm'].fillna('').str.lower().str.strip())
df_interaction.loc[df_interaction['searchTerm'].str.len() < 3, 'searchTerm'] = ''
```

```python
#handling relevancy color
df_interaction['relevancy_color'] = (df_interaction['relevancy_color'].fillna('').str.upper().replace({'': 'UNKNOWN'}))
```

```python
color_map = {'GREEN': 1, 'YELLOW': 0, 'RED': -1, 'UNKNOWN': 0}
df_interaction['relevancy_score'] = df_interaction['relevancy_color'].map(color_map)
```

```python
#handling relevancy topic
df_interaction['relevancy_topic'] = (df_interaction['relevancy_topic'].fillna('').str.lower().str.replace('_', ' ').str.strip())
```

```python
#handling state, locality, district
for col in ['state', 'locality', 'district']:
    df_interaction[col] = df_interaction[col].fillna('Unknown_Location')
```

```python
df_interaction.isnull().sum()
```

```python
df_interaction.to_csv('/content/drive/MyDrive/Companies Interview Projects/InShorts/cleaned_data/interaction_cleaned_data.csv',index=False)
```
